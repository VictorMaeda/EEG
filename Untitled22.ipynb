###Taxa de cada símbolo na fita.
    "abcdabcd" → {'ng1_a': 0.25, 'ng1_b': 0.25, ...}




```python
def freq_unigramas(fita, alphabet = ALFABETO):
    total = len(fita)
    contagem = Counter(fita)
    return {
        f"ng1_{s}": contagem.get(s, 0) / total
        for s in alphabet
    }
print(freq_unigramas(exemplo))
```

###Taxa de todos os pares de símbolos consecutivos.
    "1bcd" → {'ng2_1b': 0.33, 'ng2_bc': 0.33, 'ng2_cd': 0.33, ...resto=0}



```python
def freq_bigramas(fita, alphabet = ALFABETO):

    total = len(fita) - 1
    contagem = Counter(fita[i:i+2] for i in range(total))
    todos = ["".join(p) for p in product(alphabet, repeat=2)]
    return {
        f"ng2_{g}": contagem.get(g, 0) / total
        for g in todos
    }
print(freq_bigramas(exemplo))
```

###Taxa de todos os trios de símbolos consecutivos.
    "abcd" → {'ng3_1bc': 0.5, 'ng3_bcd': 0.5, ...resto=0}


```python
def freq_trigramas(fita, alphabet = ALFABETO):
    total = len(fita) - 2
    contagem = Counter(fita[i:i+3] for i in range(total))
    todos = ["".join(p) for p in product(alphabet, repeat=3)]
    return {
        f"ng3_{g}": contagem.get(g, 0) / total
        for g in todos
    }
print(freq_trigramas(exemplo))
```

####Entropia de Shannon de um trecho de fita, entropia de Shannon mede o grau de incerteza, aleatoriedade de uma fonte de dados.
###

```python
def _entropia_segmento(segmento):
    total = len(segmento)
    if total == 0:
        return 0.0
    contagem = Counter(segmento)
    return -sum(
        (c / total) * np.log2(c / total)
        for c in contagem.values()
        if c > 0
    )
```

###Entropia de Shannon calculada sobre a fita inteira
    - Valor alto (próximo de log2(4)=2.0) → símbolos bem distribuídos
    - Valor baixo (próximo de 0)           → fita dominada por poucos símbolos

```python
def entropia_global(fita):
    return {"entropia_global": _entropia_segmento(fita)}
print(entropia_global(exemplo))
```

####Divide a fita em janelas deslizantes e calcula a entropia de cada janela.
####Resume o resultado em 4 estatísticas (mean, std, min, max)
        mean  → nível médio de complexidade ao longo da fita
        std   → o quanto a complexidade oscila (alta std = fita heterogênea)
        min   → trecho mais "monótono" / repetitivo
        max   → trecho mais "caótico" / imprevisível

```python
def entropia_local(fita, tamanho_janela, passo = None):
    if passo is None:
        passo = tamanho_janela // 2

    entropias = [
        _entropia_segmento(fita[i: i + tamanho_janela])
        for i in range(0, len(fita) - tamanho_janela + 1, passo)
    ]

    if not entropias:
        arr = np.array([0.0])
    else:
        arr = np.array(entropias)

    w = tamanho_janela
    return {
        f"ent_local_w{w}_mean": float(arr.mean()),
        f"ent_local_w{w}_std":  float(arr.std()),
        f"ent_local_w{w}_min":  float(arr.min()),
        f"ent_local_w{w}_max":  float(arr.max()),
    }
print()
```

####Calcula o comprimento de cada "run" (bloco de símbolos iguais consecutivos)  e resume em 4 estatísticas.
        run_mean alto   → fita tende a repetir o mesmo símbolo por longos trechos
        run_mean baixo  → fita alterna símbolos rapidamente
        run_std  alto   → mistura de trechos curtos e longos
        run_max         → maior bloco repetido encontrado


```python
def runs(fita):
    lista_runs = []
    count = 1
    for i in range(1, len(fita)):
        if fita[i] == fita[i - 1]:
            count += 1
        else:
            lista_runs.append(count)
            count = 1
    lista_runs.append(count)

    arr = np.array(lista_runs, dtype=float)
    return {
        "run_mean":   float(arr.mean()),
        "run_std":    float(arr.std()),
        "run_max":    float(arr.max()),
        "run_median": float(np.median(arr)),
    }
print(runs(exemplo))
```

####Divide a fita em n_bins segmentos de igual tamanho e calcula a frequência de cada símbolo dentro de cada segmento.

#### Isso captura se o padrão dos sinais muda do início para o fim da fita (ex: símbolo 'a' predomina no começo mas some no final).
    'pos_bin0_1', 'pos_bin0_b', ...,           ← início da fita
    'pos_bin4_d',                              ← fim da fita

```python
def posicao_na_fita(
    fita,
    alphabet = ALFABETO,
    n_bins = 5
):
    tamanho = len(fita)
    tamanho_bin = tamanho // n_bins
    features = {}

    for b in range(n_bins):
        seg = fita[b * tamanho_bin: (b + 1) * tamanho_bin]
        total = len(seg)
        for s in alphabet:
            features[f"pos_bin{b}_{s}"] = seg.count(s) / total if total > 0 else 0.0

    return features
print(posicao_na_fita(exemplo))
```

```python
def tamanho_fita(fita):
    return {"tamanho_fita": len(fita)}
```

####    Extrai todas as features de uma fita e retorna um único dicionário.

        fita        : string com a sequência de sinais cerebrais
        alphabet    : lista dos 4 símbolos possíveis
        ngram_sizes : quais n-gramas calcular (1=unigrama, 2=bigrama, 3=trigrama)
        janelas     : tamanhos de janela para entropia local
        n_bins      : número de segmentos para features de posição

```python
def extrair_todas(
    fita,
    alphabet = ALFABETO,
    ngram_sizes= [1, 2, 3],
    janelas= [100, 500],
    n_bins= 5,
):
    feats = {}

    if 1 in ngram_sizes:
        feats.update(freq_unigramas(fita, alphabet))
    if 2 in ngram_sizes:
        feats.update(freq_bigramas(fita, alphabet))
    if 3 in ngram_sizes:
        feats.update(freq_trigramas(fita, alphabet))

    feats.update(entropia_global(fita))

    for w in janelas:
        if w <= len(fita):
            feats.update(entropia_local(fita, tamanho_janela=w))

    feats.update(runs(fita))
    feats.update(posicao_na_fita(fita, alphabet, n_bins))
    feats.update(tamanho_fita(fita))

    return feats
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import plotly.express as px
import pandas as pd

X = df_features.drop(columns=["label"])
y = df_features["label"]

X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

var_total = pca.explained_variance_ratio_.sum() * 100
var_pcs   = pca.explained_variance_ratio_ * 100

df_plot = pd.DataFrame({
    "PC1":   X_pca[:, 0],
    "PC2":   X_pca[:, 1],
    "PC3":   X_pca[:, 2],
    "label": y.astype(str)
})

fig = px.scatter_3d(
    df_plot,
    x="PC1", y="PC2", z="PC3",
    color="label",
    color_discrete_map={"0": "red", "1": "blue"},
    title=(
        f"PCA 3D — Variância explicada: {var_total:.2f}%<br>"
        f"PC1={var_pcs[0]:.1f}%  PC2={var_pcs[1]:.1f}%  PC3={var_pcs[2]:.1f}%"
    ),
    labels={"label": "Classe"},
    opacity=0.8
)

fig.update_traces(marker=dict(size=5))
fig.update_layout(legend_title_text="Label")

fig.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

loadings = pd.DataFrame(
    pca.components_.T,
    index=X.columns,
    columns=[f"PC{i+1}" for i in range(pca.n_components_)]
)

pesos = pca.explained_variance_ratio_
importancia = (loadings.abs() * pesos).sum(axis=1)
importancia = importancia.sort_values(ascending=False)

print("── Top 20 features mais importantes ──")
print(importancia.head(20).to_string())

plt.figure(figsize=(12, 6))
importancia.head(40).plot(kind="bar")
plt.title("Top 20 Features — Importância via PCA Loadings")
plt.ylabel("Contribuição ponderada")
plt.xlabel("Feature")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

top20_features = importancia.head(40).index
plt.figure(figsize=(10, 8))
sns.heatmap(
    loadings.loc[top20_features],
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    linewidths=0.5
)
plt.title("Loadings — Top 20 Features × Componentes Principais")
plt.tight_layout()
plt.show()

tabela = pd.DataFrame({
    "feature":     importancia.index,
    "importancia": importancia.values
}).reset_index(drop=True)
tabela
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.dummy        import DummyClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors    import KNeighborsClassifier
from sklearn.svm          import SVC
from sklearn.ensemble     import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline     import Pipeline
from sklearn.model_selection import StratifiedKFold, cross_validate
from xgboost              import XGBClassifier

X = df_features.drop(columns=["label"])
y = df_features["label"]

modelos = {
    "Dummy":         DummyClassifier(strategy="most_frequent"),
    "LDA":           LinearDiscriminantAnalysis(),
    "KNN":           KNeighborsClassifier(n_neighbors=5),
    "SVM":           SVC(kernel="rbf", C=1.0, class_weight="balanced", probability=True),
    "RandomForest":  RandomForestClassifier(n_estimators=200, class_weight="balanced", random_state=42),
    "XGBoost":       XGBClassifier(n_estimators=200, learning_rate=0.05, eval_metric="logloss", random_state=42),
}

pipelines = {
    nome: Pipeline([("scaler", StandardScaler()), ("clf", clf)])
    for nome, clf in modelos.items()
}

cv      = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = ["accuracy", "f1", "roc_auc", "precision", "recall"]
print(X.head(5))
resultados = []
for nome, pipe in pipelines.items():
    scores = cross_validate(pipe, X, y, cv=cv, scoring=scoring)
    resultados.append({
        "Modelo":    nome,
        "Accuracy":  scores["test_accuracy"].mean(),
        "F1":        scores["test_f1"].mean(),
        "ROC-AUC":   scores["test_roc_auc"].mean(),
        "Precision": scores["test_precision"].mean(),
        "Recall":    scores["test_recall"].mean(),
        "Acc_std":   scores["test_accuracy"].std(),
        "AUC_std":   scores["test_roc_auc"].std(),
    })
    print(f"✓ {nome}")

df_resultados = (
    pd.DataFrame(resultados)
    .sort_values("ROC-AUC", ascending=False)
    .reset_index(drop=True)
)

print("\n── Resultados ──")
print(df_resultados[["Modelo","Accuracy","F1","ROC-AUC","Precision","Recall"]].to_string(index=False, float_format="{:.3f}".format))

metricas = ["Accuracy", "F1", "ROC-AUC", "Precision", "Recall"]
df_plot  = df_resultados.melt(id_vars="Modelo", value_vars=metricas, var_name="Métrica", value_name="Score")

plt.figure(figsize=(13, 6))
sns.barplot(data=df_plot, x="Métrica", y="Score", hue="Modelo", palette="tab10")
plt.axhline(0.5, color="gray", linestyle="--", linewidth=0.8, label="baseline 0.5")
plt.ylim(0, 1.05)
plt.title("Comparação de Modelos — Cross-Validation (5-fold)")
plt.legend(bbox_to_anchor=(1.01, 1), loc="upper left")
plt.tight_layout()
plt.show()

df_resultados
```
